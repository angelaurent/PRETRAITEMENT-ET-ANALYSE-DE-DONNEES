**enregistrement**

le type d'enregistrement de la facture

Les types d'enregistrement EN, FC, AV et FM sur une facture SODECI en Côte d'Ivoire font référence à différentes catégories d'enregistrement de consommation d'eau. Voici leur signification :

1. EN (Enregistrement Normal) : Il s'agit de l'enregistrement de consommation d'eau basé sur la lecture régulière du compteur d'eau. L'enregistrement "EN" représente la quantité d'eau réellement consommée pendant la période de facturation.

2. FC (Facturation Complémentaire) : L'enregistrement "FC" est utilisé lorsque la lecture normale du compteur d'eau n'a pas pu être effectuée pour diverses raisons. Dans de tels cas, la SODECI estime la consommation d'eau en utilisant des méthodes alternatives, telles que des relevés antérieurs ou des estimations basées sur des paramètres spécifiques.

3. AV (Avenants) : la sodeci perçoit de l'argent sur les factures. les factures sont payes d'avance.

4. FM (Forfait Mensuel) : L'enregistrement "FM" est utilisé lorsqu'un abonné est facturé selon un forfait mensuel fixe plutôt qu'en fonction de la consommation d'eau réelle mesurée par le compteur. Cela peut être appliqué dans certaines situations spécifiques où la consommation d'eau n'est pas facilement mesurable ou lors de l'utilisation de systèmes de tarification forfaitaire.

```python
fig = plt.subplots(figsize=(8, 4))

d_enr = d1['ENR'].value_counts()
d_enr = d_enr.sort_index()

ax1 = sns.barplot(x=d_enr.index, y= d_enr.values)
plt.title("distribution des factures selon l'enregistrement")
print(d_enr)

for i, val in enumerate(d_enr.values):
    pourcentage = (val / d_enr.sum()) * 100
    ax1.text(i, val, f"{pourcentage:.1f}%", ha='center', va='bottom')


plt.ylabel('nombre de factures')
plt.xlabel('type enregistrement')

fig = plt.gcf()  # Récupère la figure actuelle
fig.savefig('D:/PROJET_TUTORE/distribution des factures selon enr.png')  # Exporte le graphe en tant qu'image PNG
```
